```python
import os
import re

# Define the folder containing your txt files
folder_path = 'iclr_parsed'  # <-- Replace with your folder path

# Define the output file path
output_file_path = 'titles.txt'  # <-- Change if you want a different output file name or location

# Regular expression pattern to fix split words caused by '#' symbols
pattern = re.compile(r'([a-zA-Z]+)#+ ?([a-zA-Z]+)')

# Open the output file for writing
with open(output_file_path, 'w', encoding='utf-8') as output_file:
    # Loop through all txt files in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith('.txt'):
            file_path = os.path.join(folder_path, filename)
            with open(file_path, 'r', encoding='utf-8') as file:
                # Read the first two lines (adjust the number if needed)
                lines = [next(file, '') for _ in range(2)]
                
                # Process each line separately
                processed_lines = []
                for line in lines:
                    # Remove leading '#' symbols
                    line = line.lstrip('#')
                    # Remove leading and trailing whitespace (including newlines)
                    line = line.strip()
                    processed_lines.append(line)
                        
                # Concatenate lines with a space
                title = ' '.join(processed_lines)
                
                # Fix split words caused by '#' symbols
                while True:
                    new_title = pattern.sub(r'\1\2', title)
                    if new_title == title:
                        break
                    title = new_title
                        
                # Remove any remaining '#' symbols
                title = title.replace('#', '')
                    
                # Replace multiple spaces with a single space and strip leading/trailing spaces
                title = re.sub(' +', ' ', title).strip()
                
                # Convert the title to lowercase
                title = title.lower()
                
                # Write the title to the output file
                output_file.write(title + '\n')
                    
                # Print the processed title to the console (optional)
                print(f"Processed title from {filename}:")
                print(title)
                print('-' * 80)
```
